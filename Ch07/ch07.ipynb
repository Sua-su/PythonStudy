```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
gdp = [67.0, 80.0 , 93.0, 123.0, 203.0, 300.0, 400.0, 600.0]

plt.plot(years, gdp, color='blue', marker='o', linestyle='solid')

plt.title("GDP per capita")

plt.ylabel("dollars")

plt.savefig("gdp_per_capita.png", dpi=600)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

x = []
y = []

for angle in range(360):
    x.append(angle)
    y.append(math.sin(math.radians(angle)))

plt.plot(x,y)
plt.title("sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.bar(range(len(years)), gdp)

plt.title("GDP per capita income")
plt.ylabel("dollars")
plt.xticks(range(len(years)),years)
plt.show()
```

```python
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [89, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x = np.arange(len(years))
plt.bar(x + 0.0, ko, width= 0.25, label="korea")
plt.bar(x + 0.3, jp, width= 0.25, label ="japan")
plt.bar(x + 0.6, ch, width=0.25, label = "china")

plt.xticks(x, years)
plt.legend()
plt.show()
```
